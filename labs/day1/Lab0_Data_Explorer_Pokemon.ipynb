### 📂 Load the Data

A CSV file is just a big table saved as text. `pd.read_csv()` loads it into a **DataFrame** — think of it as a super-powered Excel sheet inside Python.

```python
df = pd.read_csv('https://raw.githubusercontent.com/iiFadel/majal-ai-track-2026/main/labs/day1/data/Pokemon.csv')

# .head() shows the first 5 rows — always your first move!
df.head()
```

### 🔍 Section 2: Filtering — Asking Questions

Filtering means keeping only the rows that match a condition. This is how we ask the data questions!

```python
# Question: show me all the LEGENDARY Pokémon 👑
legendary = df[df['Legendary'] == True]
print(f"👑 Number of legendary Pokémon: {len(legendary)}")
legendary.head()
```

```python
# Question: show me all FIRE type Pokémon 🔥
fire = df[df['Type 1'] == 'Fire']
print(f"🔥 Number of Fire Pokémon: {len(fire)}")
fire.head()
```

```python
# You can COMBINE conditions with & (and) or | (or)
# Question: fast AND strong — Speed > 100 and Attack > 100 ⚡💪
fast_and_strong = df[(df['Speed'] > 100) & (df['Attack'] > 100)]
fast_and_strong[['Name', 'Type 1', 'Attack', 'Speed']].head(10)
```

### 🎯 Mini Exercise 2.1
Find all **Water** type Pokémon with **Attack greater than 100**. How many are there? 🌊💪

```python
# TO-DO: filter Water type Pokémon with Attack > 100
water_attack = df[(df['Type 1'] == 'Water') & (df['Attack'] > 100)]
print(f"Number of Water type Pokémon with Attack > 100: {len(water_attack)}")
```

### 🏆 Section 3: Sorting — Finding the Best

`.sort_values()` orders rows by a column. Let's settle some debates!

```python
# The 10 strongest Pokémon overall (by Total stats) 💪
strongest = df.sort_values('Total', ascending=False)
strongest[['Name', 'Type 1', 'Total', 'Attack', 'Defense', 'Speed']].head(10)
```

```python
# .value_counts() counts how many times each value appears
# Question: what is the most common Pokémon type?
df['Type 1'].value_counts()
```

```python
df['Type 2'].value_counts()
```

### 📊 Section 4: Grouping — Comparing Categories

`.groupby()` splits the data into groups and computes something for each group. Example: **which type hits hardest on average?**

```python
avg_attack = df.groupby('Type 1')['Attack'].mean().sort_values(ascending=False)
print(avg_attack)
```

```python
# Let's SEE it! Your first plot 🎨
avg_attack.plot(kind='bar', figsize=(12, 5), color='Navy')
plt.title('⚔️ Average Attack by Pokémon Type')
plt.ylabel('Average Attack')
plt.xlabel('Type')
plt.show()
```

```python
# One more: is there a relationship between Attack and Defense? 🤔
plt.figure(figsize=(8, 6))
plt.scatter(df['Attack'], df['Defense'], alpha=0.4, color='Gray')
plt.title('Attack vs Defense — every dot is a Pokémon')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show()

# 👀 Remember this kind of plot — scatter plots are the heart of Lab 1!
```
